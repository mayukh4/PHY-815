```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['grid','notebook'])
```

```python
df= pd.read_excel('GS98.xlsx')
```

```python
df.columns
```

```python
e_pp = (2.4*10**6)*((df.H1)**2)*(df.Rho)*(((df.Temp/10**6))**-0.66)*np.exp(-33.8/(df.Temp/10**6)**0.33)
```

```python
x_CNO= df.C12 + df.C13 + df.N14 + df.N15 + df.O16 + df.O17 + df.O18
```

```python
e_CNO = (8.7*10**27)*((df.H1)*(x_CNO))*(df.Rho)*(((df.Temp/10**6))**-0.66)*np.exp(-152/(df.Temp/10**6)**0.33)
```

```python
plt.plot(df.Radius, e_pp, label = '$\epsilon(pp)$')
plt.plot(df.Radius, e_CNO, label = '$\epsilon(CNO)$')
plt.title("Energy Generation rates a function of Radius", size=20)
plt.xlabel('$Radius (R/R\circ)$',size=15)
plt.ylabel('Energy generation rates ($\epsilon$) (ergs/g/s)', size=12)
plt.legend()

#plt.savefig("Energy generation rates a function of radius.jpg",dpi=500,bbox_inches='tight')
plt.show()
```
